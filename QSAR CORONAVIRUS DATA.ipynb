```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import os
```

### Searching Targets

```python
target = new_client.target
target_query = target.search('coronavirus')
target_results = pd.DataFrame.from_dict(target_query) 
```

```python
target_results
```

```python
len(target_results)
```

```python
selected_target = target_results.target_chembl_id[4]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df.head(3)
```

```python
df.standard_value
```

```python
columns = ["molecule_chembl_id","canonical_smiles","standard_value"]
df1 = df[columns]
```

```python
df1
```

```python
bioactivity_class = []
for i in df.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

```python
df1=pd.concat([df1,pd.Series(bioactivity_class,name="active_status")],axis=1)
```

```python
try:
    df1.to_csv(r'data/corona_drug1.csv',index=False)
except:
    if os.path.exists(r'data/corona_drug1.csv'):
        pass
```
