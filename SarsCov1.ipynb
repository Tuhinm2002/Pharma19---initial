```python
from chembl_webresource_client.new_client import new_client
import pandas as pd
import numpy as np
```

```python
target = new_client.target
target_query = target.search('coronavirus')
target_results = pd.DataFrame.from_dict(target_query)
```

```python
target_results
```

```python
select_target = target_results.target_chembl_id[6]
select_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=select_target).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(5)
```

```python
columns = ["molecule_chembl_id","canonical_smiles","standard_value"]
df1 = df[columns]
```

```python
df1.head(3)
```

```python
len(df1)
```

```python
df1 = df1.dropna(axis=0)
```

```python
len(df1)
```

```python
df1.standard_value.dtype
```

```python
df1['standard_value'] = np.float32(df1['standard_value'])
```

```python
df1.standard_value.dtype
```

```python
bioactivity = []
for j in df1.standard_value:
  if float(j) >= 10000:
    bioactivity.append('inactive')
  elif float(j) <= 1000:
    bioactivity.append('active')
  else:
    bioactivity.append('intermediate')
```

```python
len(bioactivity)
```

```python
df1 = pd.concat([df1,pd.Series(bioactivity,name="active_status")],axis=1)
```
