```python
df1_Y = pd.read_csv('data/corona_drug1.csv')
df1_Y.head(3)
```

```python
df2_Y = pd.read_csv('data/corona_drug2_sars2.csv')
df3_Y = pd.read_csv('data/SARS_1.csv')
```

```python
df11_Y = df1_Y['standard_value']
df22_Y = df2_Y['standard_value']
df33_Y = df3_Y['standard_value']
```

```python
df_Y = pd.DataFrame()
```

```python
df_Y = pd.concat([df_Y,df11_Y,df22_Y,df33_Y],ignore_index=True)
```

```python
len(df_Y)
```

```python
df_total
```

```python
df_Y
```

```python
bioactivity = []
for i in df_Y[0]:
    if float(i) >= 10000:
        bioactivity.append(0)
    elif float(i) <= 1000:
        bioactivity.append(1)
    else:
        bioactivity.append(2)
len(bioactivity)
```

```python
X = df_total.drop(columns='Name',axis=1)
```

```python
X
```

```python
Y = np.asarray(bioactivity)
```

```python
X.shape,Y.shape
```

```python
X_np = np.asarray(X)
```

```python
X_np.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(.8*(1-.8))
x = selection.fit_transform(X)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_np,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
# np.random.seed(42)
from sklearn.metrics import r2_score,accuracy_score
from sklearn.tree import DecisionTreeRegressor
```

```python
rfe = RandomForestClassifier(n_estimators=1000)
rfe.fit(X_train,Y_train)
r2 = rfe.score(X_test,Y_test)
r2
```

```python
y_pred = rfe.predict(X_test)
```

```python
acc = accuracy_score(Y_test,y_pred)
```
